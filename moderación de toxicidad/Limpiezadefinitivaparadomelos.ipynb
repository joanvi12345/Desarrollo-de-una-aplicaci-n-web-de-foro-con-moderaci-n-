```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords

# 1. Preparación de herramientas (Unidad 1)
nltk.download('stopwords')
stop_words_es = set(stopwords.words('spanish'))

def limpieza_profunda(texto):
    # Normalización a minúsculas
    texto = str(texto).lower() 
    
    # Quitar tildes (Simplificación de vocabulario)
    texto = re.sub(r'[áàäâ]', 'a', texto)
    texto = re.sub(r'[éèëê]', 'e', texto)
    texto = re.sub(r'[íìïî]', 'i', texto)
    texto = re.sub(r'[óòöô]', 'o', texto)
    texto = re.sub(r'[úùüû]', 'u', texto)
    
    # Quitar caracteres especiales y números (dejamos letras y espacios)
    texto = re.sub(r'[^a-zñ\s]', '', texto)
    
    # Eliminación de Stopwords (Unidad 1)
    palabras = texto.split()
    palabras_filtradas = [w for w in palabras if w not in stop_words_es]
    texto = " ".join(palabras_filtradas)
    
    # Quitar espacios múltiples
    texto = re.sub(r'\s+', ' ', texto).strip()
    return texto

# 2. Ejecución del proceso
print("Cargando DATASET_MAESTRO_BALANCEADO.csv...")
df = pd.read_csv('DATASET_MAESTRO_BALANCEADO.csv')

print("Aplicando limpieza profunda...")
df['texto_limpio'] = df['texto'].apply(limpieza_profunda)

# 3. Control de calidad (Unidad 1)
# Eliminamos filas que hayan quedado vacías tras quitar las stopwords
df = df[df['texto_limpio'] != ''].dropna(subset=['texto_limpio'])

# 4. Guardar el resultado para los modelos
archivo_salida = 'DATASET_LIMPIO_PARA_MODELOS.csv'
df.to_csv(archivo_salida, index=False, encoding='utf-8-sig')

print(f"\n--- ¡PROCESO COMPLETADO! ---")
print(f"Archivo guardado: {archivo_salida}")
print(f"Total de registros listos: {len(df)}")
print("\nComparativa de ejemplo:")
print(f"ANTES:  {df['texto'].iloc[0]}")
print(f"AHORA:  {df['texto_limpio'].iloc[0]}")
```
