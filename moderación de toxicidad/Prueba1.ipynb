```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC # Support Vector Machine del bloque 3
from sklearn.metrics import classification_report

# 1. Datos de ejemplo (Lo ideal es un CSV con miles de ejemplos)
textos = ["insulto grave", "qué buen post", "eres idiota", "saludos cordiales"]
y = [1, 0, 1, 0] 

# 2. Preprocesamiento (Unidad 1)
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(textos)

# 3. Entrenamiento (Unidad 3: SVM)
# Usamos un kernel lineal porque para texto suele ser suficiente y rápido
modelo = SVC(kernel='linear', probability=True)
modelo.fit(X, y)

# 4. Predicción de un mensaje nuevo
nuevo_mensaje = ["que buen"]  # Cambia esto por el mensaje que quieras evaluar
nuevo_X = vectorizer.transform(nuevo_mensaje)
prediccion = modelo.predict(nuevo_X)

print(f"¿Es tóxico?: {'Sí' if prediccion[0] == 1 else 'No'}")
```
