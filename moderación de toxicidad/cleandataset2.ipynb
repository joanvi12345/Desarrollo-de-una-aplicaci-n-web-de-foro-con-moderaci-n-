```python
import pandas as pd

# Tus insultos
insultos = ["Abanto", "Abrazafarolas", "Adufe", ..., "Zurcefrenillos"] # Lista completa

# Mis frases neutras inventadas
frases_neutras = [
    "Hola, ¿cómo estás?", "Me encanta aprender sobre IA.", 
    "¿Podrías ayudarme con este ejercicio?", "Hoy hace un día muy soleado.",
    "Gracias por compartir esa información.", "El respeto es la base del foro.",
    "¿Qué libro me recomiendas?", "La receta salió muy bien."
]

# Crear DataFrames
df_insultos = pd.DataFrame({'texto': insultos, 'toxicidad': 1})
df_neutros = pd.DataFrame({'texto': (frases_neutras * 30)[:len(insultos)], 'toxicidad': 0})

# Cargar el anterior y combinar
df_anterior = pd.read_csv('train_dataset_cleaned.csv')
df_final = pd.concat([df_anterior, df_insultos, df_neutros], ignore_index=True)

# Mezclar aleatoriamente
df_final = df_final.sample(frac=1).reset_index(drop=True)

# Guardar
df_final.to_csv('train_dataset_cleaned.csv', index=False, encoding='utf-8-sig')
```
