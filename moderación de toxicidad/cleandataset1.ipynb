```python
import pandas as pd
import re

# 1. Cargar el dataset original
# Si el archivo está en la misma carpeta, solo pon el nombre
df = pd.read_csv('train_dataset.csv')

# 2. Definir la expresión regular (Regex)
# Buscamos lo que está entre "Input Text:" y "<｜Assistant｜>"
# re.DOTALL permite que el buscador encuentre texto incluso si hay saltos de línea
pattern = r"<｜User｜>Input Text:\s*(.*?)\s*<｜Assistant｜>"

def extraer_mensaje(texto_sucio):
    # Buscamos el patrón en el texto de la celda
    resultado = re.search(pattern, str(texto_sucio), re.DOTALL)
    if resultado:
        # Retornamos el grupo 1 (lo que está entre paréntesis en el pattern)
        return resultado.group(1).strip()
    return None

# 3. Crear el nuevo DataFrame limpio
df_limpio = pd.DataFrame()

# Aplicamos la función a la columna 'text'
df_limpio['texto'] = df['text'].apply(extraer_mensaje)

# Asignamos toxicidad 1 a todo como pediste
df_limpio['toxicidad'] = 1

# 4. Eliminar filas que no se pudieron extraer (por si acaso hay errores en el CSV)
df_limpio = df_limpio.dropna(subset=['texto']).reset_index(drop=True)

# 5. Guardar el resultado en un nuevo CSV
df_limpio.to_csv('train_dataset_cleaned.csv', index=False, encoding='utf-8-sig')

# Ver los primeros 5 resultados
print(df_limpio.head())
print(f"\nProceso finalizado. Total de filas: {len(df_limpio)}")
```
