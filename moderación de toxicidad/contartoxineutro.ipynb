```python
import pandas as pd

# 1. Cargar el dataset definitivo
nombre_archivo = 'DATASET_MAESTRO_BALANCEADO.csv'
df = pd.read_csv(nombre_archivo)

# 2. Calcular estadísticas
total = len(df)
conteo = df['toxicidad'].value_counts()
porcentaje = df['toxicidad'].value_counts(normalize=True) * 100

# 3. Mostrar resultados por pantalla
print(f"--- ESTADÍSTICAS DE {nombre_archivo} ---")
print(f"Total de mensajes en el CSV: {total}")
print("-" * 40)
print(f"Mensajes NEUTROS (0): {conteo[0]} ({porcentaje[0]:.2f}%)")
print(f"Mensajes TÓXICOS (1): {conteo[1]} ({porcentaje[1]:.2f}%)")
print("-" * 40)

# 4. Ver una pequeña muestra de ambos para confirmar que todo está bien
print("\nEjemplo de mensaje NEUTRO:")
print(f" > {df[df['toxicidad'] == 0]['texto'].iloc[0]}")

print("\nEjemplo de mensaje TÓXICO:")
print(f" > {df[df['toxicidad'] == 1]['texto'].iloc[0]}")
```
